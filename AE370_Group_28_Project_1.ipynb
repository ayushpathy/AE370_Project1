### Mathematical Formulation of the Predator-Prey System:

The predator-prey system is governed by the **Lotka-Volterra equations**, a fundamental model in mathematical biology and engineering that describes interactions between two species: a prey population that reproduces and a predator population that depends on the prey for survival. The system is formulated as follows:

$$
\frac{dx}{dt} = \alpha x - \beta x y
$$

$$
\frac{dy}{dt} = \delta x y - \gamma y
$$

where:
- $x(t)$ represents the **prey population** at time $t$, such as rabbits in an ecosystem.
- $y(t)$ represents the **predator population** at time $t$, such as foxes that hunt the rabbits.
- $\alpha$ is the **prey birth rate**, determining how quickly the prey population grows in the absence of predators.
- $\beta$ is the **predation rate**, representing how efficiently predators consume prey.
- $\gamma$ is the **predator death rate**, defining how quickly predators die out if no prey is available.
- $\delta$ is the **conversion efficiency**, dictating how effectively consumed prey contribute to predator reproduction.

### Connecting the Model to Key Questions:

The behavior of this system is highly sensitive to the values of $\alpha, \beta, \gamma,$ and $\delta$. To address the key questions outlined earlier, we must explore different parameter values and their impact on system dynamics.

#### 2. **Do the Populations Exhibit Periodic Oscillations, Equilibrium, or Extinction?**

## ***Section 2a: Justification for Using the 4th-Order Runge-Kutta (RK4) Method***

### Accuracy

The RK4 method is a fourth-order numerical integrator, meaning that the local truncation error is on the order of $\mathcal{O}(\Delta t^5)$ and the global error is on the order of $\mathcal{O}(\Delta t^4)$. This level of accuracy allows us to:

- Accurately resolve oscillatory dynamics between predator and prey populations over time.
- Capture subtle transitions between qualitatively different behaviors (e.g., oscillations vs. extinction).
- Maintain fidelity in long-term simulations without requiring prohibitively small time steps.

This is especially important when evaluating how the system responds to small changes in parameters or initial conditions.

### Stability

Although RK4 is an explicit method and not unconditionally stable, it is sufficiently stable for the parameter ranges and time scales we are working with. The Lotka–Volterra predator-prey model is typically non-stiff under biologically realistic conditions, which makes RK4 a safe and effective choice.

- With a well-chosen step size $\Delta t$, RK4 provides stable and smooth results.
- For the initial conditions and moderate parameter ranges we are using, RK4 will not encounter significant stability issues.

If we later observe stiffness (e.g., due to extreme parameter values), we could revisit this decision and compare results using implicit methods.

### Computational Cost

RK4 requires four function evaluations per time step, which is more computationally intensive than simpler methods like Forward Euler. However:

- The method allows for a larger step size $\Delta t$ due to its higher accuracy, helping offset the cost.
- For a system of just two ODEs (prey and predator populations), the cost is modest and easily manageable on typical hardware.
- The trade-off between accuracy and performance is highly favorable for this problem.

### Summary

RK4 strikes a strong balance between accuracy, stability, and efficiency for our goals:

Therefore, RK4 is an appropriate and well-justified numerical method for exploring the dynamics of our predator-prey model.


## ***Section 2b: Mathematical Derivation of the 4th-Order Runge-Kutta (RK4) Method***

The classical RK4 method is an explicit numerical integration scheme for solving ordinary differential equations (ODEs) of the form:

$$
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
$$

To approximate the solution at discrete time steps, the RK4 method computes the value of $y$ at time $t_{n+1} = t_n + \Delta t$ using a weighted average of four derivative estimates:

### RK4 Algorithm

Given $y_n \approx y(t_n)$, compute:

$$
k_1 = f(t_n, y_n)
$$
$$
k_2 = f\left(t_n + \frac{\Delta t}{2},\ y_n + \frac{\Delta t}{2} k_1\right)
$$
$$
k_3 = f\left(t_n + \frac{\Delta t}{2},\ y_n + \frac{\Delta t}{2} k_2\right)
$$
$$
k_4 = f\left(t_n + \Delta t,\ y_n + \Delta t\, k_3\right)
$$

Then update:

$$
y_{n+1} = y_n + \frac{\Delta t}{6} \left(k_1 + 2k_2 + 2k_3 + k_4\right)
$$

Each stage estimates the slope at different points in the interval. The final update uses a weighted average to improve accuracy.

---

## Local and Global Truncation Error

- The RK4 method has a local truncation error of order:
  $$
  \mathcal{O}(\Delta t^5)
  $$
  This is the error made in a single time step.

- The global truncation error (accumulated over many steps) is:
  $$
  \mathcal{O}(\Delta t^4)
  $$
  This high-order accuracy makes RK4 an excellent choice for long-term simulations.

This means that reducing the time step by half decreases the global error by approximately a factor of 16.

---

## Stability Properties

The stability of RK4 is analyzed using the linear test equation:

$$
\frac{dy}{dt} = \lambda y, \quad y(0) = y_0
$$

Let $z = \lambda \Delta t$ be the dimensionless step-size parameter. The RK4 method applied to this equation yields the stability function:

$$
R(z) = 1 + z + \frac{z^2}{2!} + \frac{z^3}{3!} + \frac{z^4}{4!}
$$

This is the 4th-degree Taylor polynomial for $e^{z}$, which approximates the exponential function.

- The method is stable when $|R(z)| \leq 1$.
- The region in the complex plane where this condition holds defines the absolute stability region.

The RK4 method is not A-stable (i.e., not unconditionally stable for all $\text{Re}(z) < 0$), but it is sufficiently stable for non-stiff problems provided that $\Delta t$ is chosen appropriately.

---

## Summary of Properties

| Property           | Value                                     |
|--------------------|--------------------------------------------|
| Order of Accuracy  | 4th order global ($\mathcal{O}(\Delta t^4)$) |
| Local Error        | $\mathcal{O}(\Delta t^5)$                  |
| Method Type        | Explicit                                  |
| Stability          | Conditionally stable (not A-stable)       |
| Step Size Control  | Fixed step (adaptive variants exist)      |
| Recommended Use    | Smooth, non-stiff ODE systems             |

The RK4 method offers an excellent balance between computational cost, accuracy, and stability for many engineering and physics problems, including our predator-prey model under non-stiff conditions.


## ***Section 2c: Algorithmic Summary: Advancing a Solution Using RK4***

To numerically solve an ordinary differential equation (ODE) of the form:

$$
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
$$

the classical 4th-order Runge-Kutta (RK4) method advances the solution from time $( t_k $) to $$( t_{k+1} = t_k + {\Delta t} )$$ using the following steps:

---

### Inputs:
- Current time: $( t_k $)
- Current solution estimate: $( y_k $)
- Time step size:$( \Delta t $)
- Right-hand side function: $( f(t, y) $), which defines the ODE

---

### RK4 Step-by-Step:

1. Compute the first slope estimate (at the beginning of the interval):
   $$
   k_1 = f(t_k, y_k)
   $$

2. Compute the second slope estimate (at the midpoint, using \( k_1 \)):
   $$
   k_2 = f\left(t_k + \frac{\Delta t}{2},\ y_k + \frac{\Delta t}{2} \cdot k_1\right)
   $$

3. Compute the third slope estimate (again at the midpoint, using \( k_2 \)):
   $$
   k_3 = f\left(t_k + \frac{\Delta t}{2},\ y_k + \frac{\Delta t}{2} \cdot k_2\right)
   $$

4. Compute the fourth slope estimate (at the end of the interval, using \( k_3 \)):
   $$
   k_4 = f\left(t_k + \Delta t,\ y_k + \Delta t \cdot k_3\right)
   $$

5. Combine the four slopes to update the solution:
   $$
   y_{k+1} = y_k + \frac{\Delta t}{6} \left(k_1 + 2k_2 + 2k_3 + k_4\right)
   $$

6. Advance the time:
   $$
   t_{k+1} = t_k + \Delta t
   $$

---

### Outputs:
- Updated solution estimate: $( y_{k+1} $)
- Updated time: $( t_{k+1} $)

---

### Summary:
Each RK4 step uses a weighted average of four slope estimates to advance the solution one time step forward. This produces a highly accurate result without requiring the solution of algebraic equations, making RK4 ideal for non-stiff ODE systems like the Lotka–Volterra predator-prey model.

## ***3a. Numerical Simulation of the Predator-Prey System using RK4***

This notebook implements the Runge-Kutta 4th order (RK4) method to numerically solve the predator-prey (Lotka-Volterra) equations. We perform an error analysis to validate the accuracy and determine an appropriate time step for simulation.

---

### **1. Predator-Prey Model Overview**
The predator-prey system is described by the Lotka-Volterra equations:

$$
\frac{dx}{dt} = \alpha x - \beta x y
$$

$$
\frac{dy}{dt} = \delta x y - \gamma y
$$

where:
- $x(t)$ represents the prey population
- $y(t)$ represents the predator population
- $\alpha$ is the prey birth rate
- $\beta$ is the predation rate
- $\delta$ is the predator reproduction rate
- $\gamma$ is the predator death rate

These equations describe the interaction between predator and prey populations over time.

---

### **2. RK4 Method for Solving the System**
The **Runge-Kutta 4th-order method (RK4)** is a widely used explicit method for numerically solving ordinary differential equations (ODEs). It provides a good balance between accuracy and computational efficiency by using four intermediate estimates per step to compute the next values of $x$ and $y$.  

Given a system of first-order ODEs:

$$
\frac{dx}{dt} = f_x(t, x, y), \quad \frac{dy}{dt} = f_y(t, x, y),
$$

RK4 approximates the values of $x$ and $y$ at the next time step using four weighted slope estimates:

### **Step-by-step RK4 Formulation**
For each time step, we compute:

1. **First Slope Estimate ($k_1$):**
   - This is the slope at the beginning of the interval, based on the current values of $x$ and $y$.
   - It represents the instantaneous rate of change at $(t_n, x_n, y_n)$.

   $$
   k_1^x = f_x(t_n, x_n, y_n), \quad k_1^y = f_y(t_n, x_n, y_n)
   $$

2. **Second Slope Estimate ($k_2$):**
   - This estimates the slope at the **midpoint** of the interval using $k_1$ to predict new values of $x$ and $y$ at $t_n + \frac{\Delta t}{2}$.
   - It improves the estimate by incorporating the initial rate of change over half a time step.

   $$
   k_2^x = f_x\left(t_n + \frac{\Delta t}{2}, x_n + \frac{\Delta t}{2} k_1^x, y_n + \frac{\Delta t}{2} k_1^y \right)
   $$

   $$
   k_2^y = f_y\left(t_n + \frac{\Delta t}{2}, x_n + \frac{\Delta t}{2} k_1^x, y_n + \frac{\Delta t}{2} k_1^y \right)
   $$

3. **Third Slope Estimate ($k_3$):**
   - This further refines the estimate by computing another midpoint slope using $k_2$.
   - This step helps capture curvature effects in the solution.

   $$
   k_3^x = f_x\left(t_n + \frac{\Delta t}{2}, x_n + \frac{\Delta t}{2} k_2^x, y_n + \frac{\Delta t}{2} k_2^y \right)
   $$

   $$
   k_3^y = f_y\left(t_n + \frac{\Delta t}{2}, x_n + \frac{\Delta t}{2} k_2^x, y_n + \frac{\Delta t}{2} k_2^y \right)
   $$

4. **Fourth Slope Estimate ($k_4$):**
   - This calculates the slope at the **end** of the interval by using $k_3$ to estimate values at $t_n + \Delta t$.
   - It provides another crucial refinement to improve accuracy.

   $$
   k_4^x = f_x(t_n + \Delta t, x_n + \Delta t k_3^x, y_n + \Delta t k_3^y)
   $$

   $$
   k_4^y = f_y(t_n + \Delta t, x_n + \Delta t k_3^x, y_n + \Delta t k_3^y)
   $$

### **Final Update Rule**
The next values of $x$ and $y$ are obtained using a **weighted average** of the four slopes:

$$
x_{n+1} = x_n + \frac{\Delta t}{6} (k_1^x + 2k_2^x + 2k_3^x + k_4^x)
$$

$$
y_{n+1} = y_n + \frac{\Delta t}{6} (k_1^y + 2k_2^y + 2k_3^y + k_4^y)
$$

This weighted combination ensures **higher accuracy** by incorporating information from multiple points in the interval. RK4 achieves an error scaling of **$O(\Delta t^4)$**, making it much more precise than Euler’s method ($O(\Delta t)$) while maintaining good computational efficiency.

## **3. Implementation in Python**
Below, we define the model parameters, implement the RK4 solver, and perform simulations.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define model parameters
alpha = 0.5   # Prey birth rate
beta = 0.02   # Predation rate
delta = 0.01  # Predator reproduction rate
gamma = 0.3   # Predator death rate

# Define the system of equations
def predator_prey(t, x, y):
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
    return dxdt, dydt

# RK4 method implementation
def rk4_step(t, x, y, dt):
    k1x, k1y = predator_prey(t, x, y)
    k2x, k2y = predator_prey(t + dt/2, x + dt/2 * k1x, y + dt/2 * k1y)
    k3x, k3y = predator_prey(t + dt/2, x + dt/2 * k2x, y + dt/2 * k2y)
    k4x, k4y = predator_prey(t + dt, x + dt * k3x, y + dt * k3y)

    x_next = x + (dt / 6) * (k1x + 2*k2x + 2*k3x + k4x)
    y_next = y + (dt / 6) * (k1y + 2*k2y + 2*k3y + k4y)

    return x_next, y_next

# Full simulation function
def simulate_predator_prey(x0, y0, T, dt):
    t_values = np.arange(0, T, dt)
    x_values = np.zeros_like(t_values)
    y_values = np.zeros_like(t_values)

    x_values[0], y_values[0] = x0, y0

    for i in range(1, len(t_values)):
        x_values[i], y_values[i] = rk4_step(t_values[i-1], x_values[i-1], y_values[i-1], dt)

    return t_values, x_values, y_values

# Initial conditions
x0, y0 = 40, 9
T = 100  # Total time
dt = 0.0625  # Chosen time step

# Run the simulation
t_vals, x_vals, y_vals = simulate_predator_prey(x0, y0, T, dt)

# Plot results
plt.figure(figsize=(10,5))
plt.plot(t_vals, x_vals, label="Prey Population", color='blue')
plt.plot(t_vals, y_vals, label="Predator Population", color='red')
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Predator-Prey System Simulated with RK4")
plt.legend()
plt.grid()
plt.show()
```

## ***3b. Error Convergence of the RK4 Method***

### **Introduction**
To verify the expected **fourth-order accuracy** of the **Runge-Kutta 4th order (RK4)** method, we analyze how the numerical solution's error scales with respect to the time step $\Delta t$. The theoretical global error for RK4 is expected to scale as:

$$
\text{Error} \propto \mathcal{O}(\Delta t^4)
$$

This means that if we **halve** the time step $\Delta t$, the error should decrease by a factor of $2^4 = 16$.

---

### **Methodology for Computing Error Convergence**
1. **Compute a high-accuracy reference solution:**  
   - Solve the predator-prey system using RK4 with a **very small time step** ($\Delta t = 0.001$).
   - Store the reference solution $x_{\text{ref}}(t)$.

2. **Run simulations for different time steps:**  
   - Select a range of time step sizes:  
     $
     \Delta t = [1.0, 0.5, 0.25, 0.125, 0.0625]
     $
   - Solve the system using RK4 for each $\Delta t$.
   - Extract the solution $x_{\Delta t}(t)$ at matching time points.

3. **Compute the relative error:**  
   - Interpolate the numerical solution $x_{\Delta t}(t)$ to align with the high-accuracy reference time points.
   - Compute the **relative error** in the 2-norm:

$$
\text{Error} = \frac{\| x_{\Delta t} - x_{\text{ref}} \|_2}{\| x_{\text{ref}} \|_2}
$$


4. **Compare with theoretical expectations:**  
   - The errors should decrease in proportion to $\Delta t^4$.
   - We plot the error **on a log-log scale** against $\Delta t$.
   - If RK4 is behaving correctly, the slope of the error curve should be **approximately 4**.

---

### **Interpreting the Error Convergence Plot**
- **Log-log scale plot:**  
  The error values are plotted as a function of $\Delta t$ using a **log-log plot**.  
  - If RK4 is truly **fourth-order accurate**, the error plot should have a **linear trend** with a slope of approximately **4**.
  - A dashed reference line $\mathcal{O}(\Delta t^4)$ is plotted for comparison.

- **Confirming the expected behavior:**  
  - If the **measured errors** align with the expected error curve $\mathcal{O}(\Delta t^4)$, this confirms that RK4 is behaving as expected.
  - If deviations occur, they may be due to numerical instability, insufficient reference accuracy, or floating-point precision limits.

---

### **Conclusion**
This method systematically verifies the error convergence rate of RK4 by comparing numerical solutions at different time steps to a high-accuracy reference solution. The expected **fourth-order scaling** should be clearly visible in the log-log plot. If the RK4 method is correctly implemented, the **error curve should follow a slope of 4**, confirming the method's accuracy.


```python
# Compute RK4 error convergence
dt_values = [1.0, 0.5, 0.25, 0.125, 0.0625]  # Different time step sizes
errors = []

# Use a reference solution with very small dt
t_ref, x_ref, y_ref = simulate_predator_prey(40, 9, 100, 0.001)
x_ref_interp = np.interp(np.arange(0, 100, 0.001), t_ref, x_ref)

for dt in dt_values:
    t_vals, x_vals, y_vals = simulate_predator_prey(40, 9, 100, dt)
    x_interp = np.interp(np.arange(0, 100, 0.001), t_vals, x_vals)
    error = np.linalg.norm(x_interp - x_ref_interp, ord=2) / np.linalg.norm(x_ref_interp, ord=2)
    errors.append(error)

# Expected theoretical error scaling O(dt^4)
dt_values = np.array(dt_values)
expected_errors = errors[-1] * (dt_values / dt_values[-1])**4  # Scale the smallest dt error by O(dt^4)

# Generate error convergence plot
plt.figure(figsize=(8, 6))
plt.loglog(dt_values, errors, 'o-', color='orange', label="RK4 Error")
plt.loglog(dt_values, expected_errors, '--', color='brown', label=r"Expected $\mathcal{O}(\Delta t^4)$")
plt.xlabel("Time Step $\Delta t$")
plt.ylabel("Relative Error")
plt.title("Error Convergence of RK4 Method")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

## ***3c. Necessary Simulation Parameters***

### Simulation Parameters for Accurate Predator-Prey Dynamics Study

To accurately simulate and study the predator-prey system using the RK4 method, we need to carefully choose key simulation parameters. These parameters include the time step size ($\Delta t$), the total simulation time ($T$), and the error convergence analysis. Here are the key factors to consider:

#### 1. **Time Step Size ($\Delta t$)**
   - The **time step size** plays a crucial role in determining the accuracy and computational cost of the simulation. The RK4 method has a **global error of order 4**, meaning that the error decreases at a rate of $O(\Delta t^4)$ as the time step size ($\Delta t$) is reduced.
   - **Smaller time steps** increase accuracy but also increase computational cost. Therefore, we must find a balance between accurate results and efficient computation.
   - A common approach is to start with a moderate time step, such as $\Delta t = 0.1$ or $\Delta t = 0.05$, and then refine the step size based on the results and error convergence study.

#### 2. **Total Simulation Time ($T$)**
   - For long-term simulations (large $T$), smaller time steps are often necessary to prevent numerical errors from accumulating and distorting the solution.
   - If the system exhibits **periodic behavior** (e.g., predator-prey cycles), the time step should be chosen small enough to capture these oscillations accurately. A good rule of thumb is to have at least **10–20 time steps per oscillation period**. If the predator-prey system exhibits periodic oscillations with period $T_{\text{period}}$, the time step should satisfy:
     $$
     \Delta t \leq \frac{T_{\text{period}}}{20}
     $$
   - For example, if the period of oscillation is $T_{\text{period}} = 20$, you might choose $\Delta t \leq 1$ to capture the dynamics effectively.

#### 3. **Error Convergence**
   - We can use an **error convergence study** to determine the optimal time step. This involves comparing the solution obtained with a coarse time step to that of a fine time step and evaluating the error.
   - The error should be calculated using a **norm-based error function**, such as the Euclidean norm, and then normalized by the norm of the fine solution. We expect the error to converge as $O(\Delta t^4)$ for RK4.
   - The error convergence is visualized by plotting the relative error against the time step size on a log-log scale. The plot should show a slope of approximately $-4$ for the RK4 method.

#### 4. **Stability Considerations**
   - The RK4 method is **generally stable** for most well-behaved ODEs, but if the time step is too large, the system may become unstable. This can result in wild oscillations or unbounded growth in the populations, especially in stiff systems or systems with rapid changes.
   - If the predator-prey system shows erratic behavior or unrealistic population sizes, consider decreasing the time step size.

#### 5. **Empirical Testing and Sensitivity**
   - To find an appropriate time step, it's useful to run simulations for different time step values (e.g., $\Delta t = 0.5, 0.25, 0.125, 0.0625$) and compare the results. You can check the behavior of the populations over time and monitor the error.
   - Perform a **convergence study**:
     - Run simulations with different time step sizes.
     - Compare the error in predator and prey populations.
     - Plot the error as a function of time step size to observe the convergence behavior.

#### 6. **Guidelines for Simulation**
   - **Start with moderate time steps** (e.g., $\Delta t = 0.1$ or $\Delta t = 0.05$) and monitor the accuracy of the solution.
   - **Refine the time step** (e.g., to $\Delta t = 0.01$ or smaller) if finer accuracy is needed in regions of rapid change or oscillations.
   - **Check for instability**: If the populations show unrealistic behavior (e.g., extreme oscillations or explosions), reduce the time step size further.

#### 7. **Final Recommendations**
   - For long-term simulations ($T = 100$), starting with $\Delta t = 0.1$ or $\Delta t = 0.05$ is reasonable. If accuracy is acceptable, maintain the time step at these values. Otherwise, reduce $\Delta t$ and check the error.
   - For simulations of shorter durations, larger time steps (e.g., $\Delta t = 0.25$ or $\Delta t = 0.125$) may be sufficient.

By considering these factors and conducting an error convergence study, you can select the appropriate time step size to ensure an accurate and efficient simulation of the predator-prey system.
